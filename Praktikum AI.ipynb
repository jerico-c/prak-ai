```python
from collections import deque

graph = {
    'Secang' : ['Kota Magelang'],
    'Kota Magelang' : ['Mertoyudan', 'Tegalrejo'],
    'Tegalrejo' : [],
    'Mertoyudan' : ['Mungkid'],
    'Mungkid' : ['Muntilan', 'Sawangan'],
    'Sawangan' : [],
    'Muntilan' : ['Salam'],
    'Salam' : ['Ngluwar'],
    'Ngluwar' : [],
}

def bfs(graph, node) :
    visited = []
    queue = deque()

    visited.append(node)
    queue.append(node)

    while queue:

      q = queue.popleft()
      print(q, end = " ")

      for n in graph[q]:
        if n not in visited:
          visited.append(n)
          queue.append(n)

def main():
    bfs(graph, 'Secang')

main()
```
