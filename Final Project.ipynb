```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

# Muat dataset Iris
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

# Pisahkan dataset menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Buat classifier Gaussian Naive Bayes
gnb = GaussianNB()

# Latih classifier
gnb.fit(X_train, y_train)

# Prediksi label set pengujian
y_pred = gnb.predict(X_test)

# Hitung akurasi classifier
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi: {accuracy:.2f}')

# Plot jumlah data per kelas
unique, counts = np.unique(y, return_counts=True)
plt.bar(target_names, counts, color=['red', 'yellow', 'green'])
plt.xlabel('Kelas')
plt.ylabel('Jumlah Data')
plt.title('Jumlah Data per Kelas')
plt.show()

# Visualisasi data: Sepal Length vs Sepal Width
plt.figure(figsize=(10, 6))
colors = ['red', 'yellow', 'green']
for i, color in enumerate(colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, label=target_names[i], edgecolor='k')

plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.title('Sepal Length vs Sepal Width')
plt.show()
```
